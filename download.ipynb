```python
from pathlib import Path
import json

json_path = Path("msrvtt_train_7k.json")
video_dir = Path("data/1/TrainValVideo")

records = json.loads(json_path.read_text())
json_videos = {item["video"] for item in records}
dataset_videos = {p.name for p in video_dir.glob("*.mp4")}

covered = dataset_videos & json_videos
missing = sorted(dataset_videos - json_videos)

print(f"dataset videos: {len(dataset_videos)}")
print(f"json videos:    {len(json_videos)}")
print(f"covered:        {len(covered)} ({len(covered)/len(dataset_videos):.2%})")
print(f"missing names:  {missing[:10]}")
```
